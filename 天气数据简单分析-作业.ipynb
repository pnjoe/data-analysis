### 绘图(折线图)

看到上图是不是很混乱.所有数据都夹杂上去.单位也不统一.
于是乎,我就来单项输出分析吧 ^_^

```python
# 全年温度折线图 X轴:日期  Y轴:温度  单位:华氏温标 F
weather.plot(y='TemperatureF')
```

华氏温度的数据对我来说不是很敏感.平常用的是摄氏温度 ℃ .
 那么可否把它转换成摄氏温度呢?

有了想法.那就行动吧.

谷歌了一下.找到了换算公式.
华氏温度F 换算成 摄氏温度℃：
### ℃ = (F - 32) / 1.8

因小白,暂时不会 在不动原数据框数据的前提下,运用公式直接换算 并在图表上呈现.
 但课堂上老师有教新增列的函数.那我不是可以新增一列摄氏温度的数据.
然后再图表展现这一列的数据呢.
我来试一下吧

```python
# 新增 Celsius 列.  
# 生成新的摄氏温度加到这张表格里.
weather['Celsius'] = ( weather['TemperatureF'] - 32 ) / 1.8
```

```python
weather.plot(y='Celsius')
```

现在再看这图表.是否更清晰了呢.

图表上可以明显看出6-8月份温度高(夏季) 12,1,2月份温度低(冬季) 

可以推断出是北半球地区. [南半球季节划分是怎么样？]('https://zhidao.baidu.com/question/569676865.html')

冬季会下雪,  ... 可以推断出此地区是 在[北温带]('http://baike.baidu.com/item/北温带') 

具体是什么[气候类型]('http://baike.baidu.com/item/气候类型')估计需要参温度才好判断吧.


```python
#查下全年最高温度是多少?  具体是哪一天?(不会写函数,求老师指点)
max(weather['Celsius'])
```
